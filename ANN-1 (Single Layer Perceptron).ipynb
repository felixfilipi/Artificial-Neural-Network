```python
import numpy as np #inport numpy with alias np

# preapre dataset

# 1 XOR 1 ==> 0
# 1 XOR 0 ==> 1
# 0 XOR 1 ==> 1
# 0 XOR 0 ==> 0

dataset = [[1,1], [1,0], [0,1], [0,0]]
target = [0, 1, 1, 0]

# neuron model
layer = {
    "input": 2,
    "output": 1
}

# initial weight and bias
# stochastic (random dataset)
weight = np.random.rand(layer["input"])
bias = np.random.rand(layer["output"])

#print(np.array(dataset).shape) // print dataset's array
#print(np.array(target).shape) // print target's array 

# feed forward (process)
def feed_forward(dataset):
    u = np.matmul(dataset, weight)
    return activation(u+bias)

# activation function
def activation(val):
    if(val >= 0):
        return 1
    else:
        return 0

# set epoch and learning rate
epoch = 10 #training loop
learning_rate = 0.2 #training speed

for i in range(epoch):
    rand = np.random.randint(0,4)
    data = dataset[rand]

    # get output
    output = feed_forward(data)
    print(output)

    # calculate error
    error = target[rand] - output

    # update weight and bias
    weight = np.add(weight, np.array(data) * learning_rate * error)
    bias = bias + learning_rate * error

    print(weight)
    print(bias)

matches = 0
for index, data in enumerate(dataset):
    output = feed_forward(data)
    print("Expectation: ", target[index])
    print("Result: ", output)
    if(output == target[index]):
        matches += 1

print("Accuracy {}%".format(matches*100/4))
```
